```python
from pathlib import Path
from utils.ops import load_yaml, save_yaml
```

```python
exps_path = Path('experiments')
```

```python
for t in exps_path.rglob('train_results.yaml'):
    train_results = load_yaml(t)
    model_path = str(train_results['model_path'])
    model_path = Path(model_path)
    train_results['model_file'] = model_path.name
    print(model_path.name)
    save_yaml(train_results, t)
    print(t)
```
